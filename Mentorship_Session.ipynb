Use the Cleaned data to Award the Points As per the provided rules.
The Mentee_ID column was adjusted using Excel.

```python
import pandas as pd

# Load the cleaned data
file_path = '/content/cleaned_data.xlsx'
df = pd.read_excel(file_path, sheet_name='Mentorship_Sessions')


# Initialize the point_awards column
df['Point_Awards'] = 0
# Rule 1: Award 250 points for signing up as a mentor (one-time allocation)
df['Point_Awards'] += 250
# Rule 2: Award 1000 points for conducting mentorship with 2 different mentees
# Count the number of unique mentees for each mentor
mentee_counts = df.groupby('Mentor_ID')['Mentee_ID'].nunique()

# Apply the rule to the dataframe
df['Point_Awards'] += df['Mentor_ID'].map(lambda mentor: 1000 if mentee_counts[mentor] >= 2 else 0)

# Rule 3: Allocate points based on session conditions
# Filter rows where session length is at least 30 minutes
valid_sessions = df[df['Session_Duration_Min'] >= 30]

# Check if there are two sessions with the same mentee for each mentor
session_counts = valid_sessions.groupby(['Mentor_ID', 'Mentee_ID']).size()

# Apply points for mentorship relationships
df['Point_Awards'] += df.apply(lambda row: 500 if session_counts.get((row['Mentor_ID'], row['Mentee_ID']), 0) >= 2 else 250 if row['Session_Duration_Min'] >= 30 else 0, axis=1)

# Rule 4: Award additional 250 points for sessions with job information
df['Point_Awards'] += df.apply(lambda row: 250 if row['Job_Info_Completed'] == 'Yes' else 0, axis=1)

# # Save the updated dataframe
# df.to_excel('/mnt/data/updated_mentorship_sessions_with_points.xlsx', index=False)

# df.head()  # Display the first few rows to verify
```
